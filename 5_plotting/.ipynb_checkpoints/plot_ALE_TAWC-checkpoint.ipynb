# This script generates the data for ALE plots

```python
import glob
import pandas as pd
import numpy as np

from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import roc_curve, auc
from sklearn.base import clone 

from scipy.stats import ks_2samp
from hyperopt import hp, fmin, tpe, STATUS_OK, Trials
from functools import partial

from tqdm.notebook import trange, tqdm

import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
from PyALE import ale

from pathlib import Path

import warnings
warnings.filterwarnings("ignore")
```

```python
# Set paths
ROOT = Path('/gpfs/work1/0/wtrcycle/users/jaerts/model_refinement_pub/')
RESULTS = f'{ROOT}/results/'
AUXDIR = f'{ROOT}/aux_data/'
# NOTE this is TEMPORARY
OUTPUT = f'/gpfs/home6/jaerts/temp_model_refinement/rf_feature_selection/ale_data/'
```

```python
# Load Streamflow Dataframes
df_streamflow = pd.read_csv(f'{RESULTS}/rf_feature_selection/rf_wflow_pcrglob_streamflow_camels_hydro_attrs.csv', index_col='basin_id')
df_streamflow.loc[df_streamflow['kge_np_wflow'] < -0.42, 'kge_np_wflow'] = -0.42
df_streamflow.loc[df_streamflow['kge_np_pcrglob'] < -0.42, 'kge_np_pcrglob'] = -0.42
df_streamflow['kge_np_delta_wp'] = df_streamflow.kge_np_wflow - df_streamflow.kge_np_pcrglob
df_streamflow = df_streamflow.dropna()

# Load Evaporation Dataframes
df_evaporation = pd.read_csv(f'{RESULTS}/rf_feature_selection/rf_wflow_pcrglob_evaporation_camels_hydro_attrs.csv', index_col='basin_id')
df_evaporation['RRMSE_et_delta_wp'] = df_evaporation.RRMSE_et_wflow - df_evaporation.RRMSE_et_pcrglob
df_evaporation = df_evaporation.dropna()

# Load Soil Moisture Dataframes
df_soilmoisture = pd.read_csv(f'{RESULTS}/rf_feature_selection/rf_wflow_pcrglob_soil_moisture_camels_hydro_attrs.csv', index_col='basin_id')
df_soilmoisture.loc[df_soilmoisture['RRMSE_sm_wflow'] > 10, 'RRMSE_sm_wflow'] = 10
df_soilmoisture.loc[df_soilmoisture['RRMSE_sm_pcrglob'] > 10, 'RRMSE_sm_pcrglob'] = 10
df_soilmoisture['RRMSE_sm_delta_wp'] = df_soilmoisture.RRMSE_sm_wflow - df_soilmoisture.RRMSE_sm_pcrglob
df_soilmoisture = df_soilmoisture.dropna()
```

```python
# Load abstraction level
df_abs = pd.read_csv(f'{AUXDIR}/rf_features/independent_variable_abstraction_level.csv', index_col='Features')

# Streamflow
variable = 'streamflow'
imp_wflow_q = pd.read_csv(f'{RESULTS}/rf_feature_selection/feature_importance/feature_importance_rf_wflow_{variable}.csv', index_col='Feature')
imp_wflow_q = imp_wflow_q.join(df_abs)
imp_wflow_q = imp_wflow_q.sort_values('Importance',ascending=False)

imp_pcrglob_q = pd.read_csv(f'{RESULTS}/rf_feature_selection/feature_importance/feature_importance_rf_pcrglob_{variable}.csv', index_col='Feature')
imp_pcrglob_q = imp_pcrglob_q.join(df_abs)
imp_pcrglob_q = imp_pcrglob_q.sort_values('Importance',ascending=False)

imp_wp_q = pd.read_csv(f'{RESULTS}/rf_feature_selection/feature_importance/feature_importance_rf_delta_wp_{variable}.csv', index_col='Feature')
imp_wp_q = imp_wp_q.join(df_abs)
imp_wp_q = imp_wp_q.sort_values('Importance',ascending=False)

# Evaporation
variable = 'evaporation'
imp_wflow_et = pd.read_csv(f'{RESULTS}/rf_feature_selection/feature_importance/feature_importance_rf_wflow_{variable}.csv', index_col='Feature')
imp_wflow_et = imp_wflow_et.join(df_abs)
imp_wflow_et = imp_wflow_et.sort_values('Importance',ascending=False)

imp_pcrglob_et = pd.read_csv(f'{RESULTS}/rf_feature_selection/feature_importance/feature_importance_rf_pcrglob_{variable}.csv', index_col='Feature')
imp_pcrglob_et = imp_pcrglob_et.join(df_abs)
imp_pcrglob_et = imp_pcrglob_et.sort_values('Importance',ascending=False)

imp_wp_et = pd.read_csv(f'{RESULTS}/rf_feature_selection/feature_importance/feature_importance_rf_delta_wp_{variable}.csv', index_col='Feature')
imp_wp_et = imp_wp_et.join(df_abs)
imp_wp_et = imp_wp_et.sort_values('Importance',ascending=False)

# Soil Moisture
variable = 'soil_moisture'
imp_wflow_sm = pd.read_csv(f'{RESULTS}/rf_feature_selection/feature_importance/feature_importance_rf_wflow_{variable}.csv', index_col='Feature')
imp_wflow_sm = imp_wflow_sm.join(df_abs)
imp_wflow_sm = imp_wflow_sm.sort_values('Importance',ascending=False)

imp_pcrglob_sm = pd.read_csv(f'{RESULTS}/rf_feature_selection/feature_importance/feature_importance_rf_pcrglob_{variable}.csv', index_col='Feature')
imp_pcrglob_sm = imp_pcrglob_sm.join(df_abs)
imp_pcrglob_sm = imp_pcrglob_sm.sort_values('Importance',ascending=False)

imp_wp_sm = pd.read_csv(f'{RESULTS}/rf_feature_selection/feature_importance/feature_importance_rf_delta_wp_{variable}.csv', index_col='Feature')
imp_wp_sm = imp_wp_sm.join(df_abs)
imp_wp_sm = imp_wp_sm.sort_values('Importance',ascending=False)
```

```python
# Figure settings
matplotlib.rcParams.update({'font.size': 18})
sns.set_style("whitegrid")
sns.color_palette("tab10")
colors = {'meteorology': 'tab:blue',
          'hydrology': 'tab:orange',
          'land surface': 'tab:green',
          'subsurface': 'tab:red'}

fig, axes = plt.subplots(1,3, figsize=(20,5))

# Plot three selected features (area, low_prec_freq, tawc)
features = ['tawc','area','area','low_prec_freq']
###################################################################################
# FIRST ROW
# Plot first feature Streamflow
variable = 'streamflow'
feature = features[0]
print(variable,feature)
imp_1 = imp_wflow_q.loc[feature]['Importance'] 
imp_2 = imp_pcrglob_q.loc[feature]['Importance'] 
imp_3 = imp_wp_q.loc[feature]['Importance'] 

file_1 = glob.glob(f'{OUTPUT}/{variable}/wflow/*_{feature}_*')[0]
file_2 = glob.glob(f'{OUTPUT}/{variable}/pcrglob/*_{feature}_*')[0]
file_3 = glob.glob(f'{OUTPUT}/{variable}/delta_wp/*_{feature}_*')[2]

df_1 = pd.read_csv(file_1)
df_2 =pd.read_csv(file_2)
df_3 = pd.read_csv(file_3)

# Bin data
bins = 20
df_1 = pd.DataFrame({"y": df_1.eff,
                  "x": df_1[feature],
                  "bin_code": pd.cut(df_1[feature], bins, include_lowest=True).cat.codes + 1,})
df_2 = pd.DataFrame({"y": df_2.eff,
                  "x": df_2[feature],
                  "bin_code": pd.cut(df_2[feature], bins, include_lowest=True).cat.codes + 1,})
df_3 = pd.DataFrame({"y": df_3.eff,
                  "x": df_3[feature],
                  "bin_code": pd.cut(df_3[feature], bins, include_lowest=True).cat.codes + 1,})

# Plot
ax1 = sns.scatterplot(ax=axes[0], data=df_1, y="y", x="x", color='tab:red')
ax2 = sns.scatterplot(ax=axes[1], data=df_2, y="y", x="x", color='tab:red')
ax3 = sns.scatterplot(ax=axes[2], data=df_3, y="y", x="x", color='tab:red')

ax1.axvline(x = 115, color = 'black', linestyle='dashed',linewidth=2)
ax2.axvline(x = 95, color = 'black', linestyle='dashed',linewidth=2)
# ax3.axvline(x = 7, color = 'b', label = 'axvline - full height')

# Center around zero
yabs_max = abs(max(ax1.get_ylim(), key=abs))
yabs_max = 0.03
ax1.set_ylim(ymin=-yabs_max, ymax=yabs_max)
ax2.set_ylim(ymin=-yabs_max, ymax=yabs_max)
ax3.set_ylim(ymin=-yabs_max, ymax=yabs_max)

# Set labels
feature = "TAWC (mm)"
abs_class = "Subsurface"
ax1.set_xlabel(f'{feature}')
ax2.set_xlabel(f'{feature}')
ax3.set_xlabel(f'{feature}')

ax1.set_title(f'RFM-wflow_sbm: Streamflow')
ax2.set_title(f'RFM-PCR-GLOBWB: Streamflow')
ax3.set_title(f'RFM-$\Delta$(W-P): Streamflow')

ax1.set_ylabel(f'{abs_class} class \n Effect on Prediction')
ax2.set_ylabel(None)
ax3.set_ylabel(None)

# # # Add attonation
import string

axes[0].text(-0.15,1.05, string.ascii_uppercase[0], transform=axes[0].transAxes, size=30, weight='bold')
axes[1].text(-0.15,1.05, string.ascii_uppercase[1], transform=axes[1].transAxes, size=30, weight='bold')
axes[2].text(-0.15,1.05, string.ascii_uppercase[2], transform=axes[2].transAxes, size=30, weight='bold')

# # # just align the last column of axes:

plt.tight_layout()
# fig.align_ylabels(axes[:, 1])
plt.savefig(f'/home/jaerts/model_refinement_master/new_scripts/figures/7_ALE_TAWC_example.png', bbox_inches='tight', dpi=500)
```
