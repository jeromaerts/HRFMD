# Prepare input data for RF 

```python
import pandas as pd
from pathlib import Path
```

## Set Paths

```python
ROOT = Path('/gpfs/work1/0/wtrcycle/users/jaerts/model_refinement_pub/')
ATTRDIR = f'{ROOT}/catchment_attributes/'
RESULTS = f'{ROOT}/results/'
```

## Combine independent and dependent variables

```python
# Load CAMELS attrs only
df_camels = pd.read_csv(f'{RESULTS}/new/rf_feature_selection/camels_attributes_hydrological_signatures_rf_new.csv',index_col='gauge_id')
```

### Streamflow wflow & pcrglobwb CAMELS attrs

```python
# Load streamflow results
df_wflow = pd.read_csv(f'{RESULTS}/streamflow_evaluation/wflow_sbm_evaluation/wflow_streamflow_evaluation_objective_functions_overview.csv', index_col='basin_id')
df_pcrglob = pd.read_csv(f'{RESULTS}/streamflow_evaluation/pcr-globwb_evaluation/pcr-globwb_streamflow_evaluation_objective_functions_overview.csv', index_col='basin_id')
```

```python
df_wflow = df_wflow['kge_np'].to_frame()
df_pcrglob = df_pcrglob['kge_np'].to_frame()

df = df_wflow.join(df_pcrglob, lsuffix='_wflow', rsuffix='_pcrglob')
df = df.join(df_camels)
df = df.dropna()
```

```python
# write output
df.to_csv(f'{RESULTS}/new/rf_feature_selection/rf_wflow_pcrglob_streamflow_camels_attrs.csv')
```
